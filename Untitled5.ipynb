```python
import pandas as pd
import json

def process_csv(file_path):
    df = pd.read_csv(file_path)
    predicted_priority = []
    predicted_resolution_time = []

    for index, row in df.iterrows():
        bug_description = row['input_text']
        prediction = classify_bug(bug_description)
        print(prediction,"\n\n")

        # Extract predictions from the JSON-like response
        try:
            prediction_dict = json.loads(prediction)  # Ensure it's proper JSON format
            priority = prediction_dict.get("Priority", "unknown")
            resolution_time = prediction_dict.get("Resolution Time", "unknown")
        except json.JSONDecodeError:
            priority = "unknown"
            resolution_time = "unknown"

        predicted_priority.append(priority)
        predicted_resolution_time.append(resolution_time)

    df['predicted_priority'] = predicted_priority
    df['predicted_resolution_time'] = predicted_resolution_time

    # Save predictions to CSV
    output_file = "predicted_bugs_llama.csv"
    df.to_csv(output_file, index=False)
    print(f"Predictions saved to {output_file}")

# Example usage: 
file_path = "updated_combined_data.csv"  # Update this with the actual file path
process_csv(file_path)
```
